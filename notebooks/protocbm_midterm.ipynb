# Plots for the midterm presentation

```python
import numpy as numpy
from matplotlib import pyplot as plt
```

## Number of prototype vectors

```python
num_prototypes = range(1, 11)
class_accuracy = [77.80, 77.88, 80.22, 78.80, 78.21, 77.55, 79.22, 79.38, 80.97, 79.80]
attribute_accuracy = [77.59, 80.63, 80.72, 79.96, 80.31, 80.85, 81.73, 80.45, 81.05, 80.82]
loc_distance = [58.44, 65.13, 84.96, 72.69, 65.51, 80.47, 72.68, 83.69, 67.25, 71.85]

# Plot all in one plot
plt.plot(num_prototypes, class_accuracy, label='Class Accuracy')
plt.plot(num_prototypes, attribute_accuracy, label='Attribute Accuracy')
plt.plot(num_prototypes, loc_distance, label='Location Distance')
plt.xlabel('Number of Prototype Vectors')
plt.ylabel('Metrics')

# x-axis should only show integers
plt.xticks(num_prototypes)

# Add a trendline for each metric
z3 = numpy.polyfit(num_prototypes, loc_distance, 1)
p3 = numpy.poly1d(z3)
plt.plot(num_prototypes, p3(num_prototypes), "b--", label='Location Distance Trendline', color='green')
plt.legend()
plt.show()
```
